```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
url='/content/E_shop_cleaned_data.csv'
df=pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
#Unfortunetly, a data set description is not avaiable for these data on UCI website. So, I have to figure out the type of the variables by checking their values
#For now, I draw histogram charts for all the numeric variables to see which ones are actually numeric and which one are actually categorical presented by numbers
```

```python
#Distributions for all the numeric variables
```

```python
num=df[['Year','Month','Day','Order','Country','Session ID','Page 1 (main category)','Color','Location','Model photography','Price','Price 2','Page']]
for i in num.columns:
  plt.hist(num[i])
  plt.title(i)
  plt.show()
```

```python
#There is only one value for variable Year which is 2008. So this variable has no effect on the model and there is no need to use it in the model building part
#Some variables are considered by Python to be numeric, but they are actually categorical. These variables are:
#Month and Day (changing in them doesn't mean increase or decrease, it just means another month or day)
#Country is presnted by numbers in this data set, but it is a categorical variable. This is also true for variable Location
#Color is also like Country
#Model photography is a binary variable and so categorical. This is also true for Price 2
#Page 1 should be categorical too
#Page is also categorical variables
```

```python
df['Month']=df['Month'].astype('object')
df['Day']=df['Day'].astype('object')
df['Country']=df['Country'].astype('object')
df['Location']=df['Location'].astype('object')
df['Color']=df['Color'].astype('object')
df['Model photography']=df['Model photography'].astype('object')
df['Price 2']=df['Price 2'].astype('object')
df['Page 1 (main category)']=df['Page 1 (main category)'].astype('object')
df['Page']=df['Page'].astype('object')
```

```python
############ Clustering models
```

```python
X=df[['Month','Day','Country','Page 1 (main category)','Page 2 (clothing model)','Color','Location','Model photography','Price 2','Page','Order','Session ID','Price']]
X=pd.get_dummies(X)
```

```python
### K-means model
#First, I have to find the optimized number of clusters. I use elbow method for that
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++')
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
```

```python
#Based on the plot, the best number of clusters is 3. So, I make the K-means model with k = 3
#I define variable y_kmeans that is obtained by K-means algorithm and determine the cluster of each observation
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
y_kmeans=kmeans.fit_predict(X)
```

```python
#I can add this new variable which shows the cluster of each observation using k-means algorithm to the dataframe
df['Cluster_Kmeans']=y_kmeans
```

```python
df['Cluster_Kmeans'].value_counts()
```

```python
#It is clear how many observations are in each category
```

```python
### Hierarchical model
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
#I can also add this new variable which shows the cluster of each observation using hierarchical algorithm to the dataframe
df['Cluster_HC']=y_hc
```
